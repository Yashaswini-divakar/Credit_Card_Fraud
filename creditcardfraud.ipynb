```python
## Independent and dependent features
x = df.iloc[:,:-1]
print(x)
y = df.iloc[:, -1]
print(y)
```

###Cross Validation: KFold and Hyperparameter Technique: GridSearchCV

```python
import numpy as np
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import KFold
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import train_test_split
```

```python
log_reg = LogisticRegression()
grid = { 'solver': ['liblinear', 'saga'] , "C" : 10.0 ** np.arange(-2,3) , "penalty" : ["l1" , "l2"]}
cv = KFold(n_splits = 5 , random_state = None , shuffle = False )
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X = scaler.fit_transform(x)
```

```python
x_train , x_test , y_train , y_test = train_test_split(X,y,test_size=0.25)
print(x_train.shape , y_train.shape)
print(x_test.shape , y_test.shape)
```

```python
clf = GridSearchCV(log_reg , grid, cv =cv , n_jobs=-1 , scoring = "f1_macro")
clf.fit(x_train , y_train)
```

```python
y_pred = clf.predict(x_test)
print(accuracy_score(y_test , y_pred))
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test , y_pred))
```

```python
class_weigh = dict({ 0 : 1 , 1 : 100})
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier(class_weight = class_weigh)
rfc.fit(x_train,y_train)
```

```python
y_pred1 = rfc.predict(x_test)
print(accuracy_score(y_test,y_pred1))
print(confusion_matrix(y_test,y_pred1))
print(classification_report(y_test , y_pred1))
```

### UNDER SAMPLING

```python
from imblearn.under_sampling import NearMiss
from collections import Counter
```

```python
ns = NearMiss(sampling_strategy=0.8)
x_train_ns , y_train_ns = ns.fit_resample(x_train , y_train)
print("The number of y_train cases before fit {} ".format(Counter(y_train)))
print("The number of y_train_ns cases after fit {} ".format(Counter(y_train_ns)))
```

```python
print(y_train.value_counts())
print(y_train_ns.value_counts())
```

```python
ns_rfc = RandomForestClassifier()
ns_rfc.fit(x_train_ns , y_train_ns)
y_pred_ns = ns_rfc.predict(x_test)
```

```python
print(accuracy_score(y_test , y_pred_ns))
print(confusion_matrix(y_test , y_pred_ns))
print(classification_report(y_test , y_pred_ns))
```

### OVER SAMPLING

```python
from imblearn.over_sampling import RandomOverSampler
```

```python
os = RandomOverSampler(sampling_strategy = 0.5)
x_train_os , y_train_os = os.fit_resample(x_train , y_train)
print("The number of classes before fit {} ".format(Counter(y_train)))
print("The number of classes after fit {} ".format(Counter(y_train_os)))
```

```python
print(y_train.value_counts())
print(y_train_os.value_counts())
```

```python
os_rfc = RandomForestClassifier()
os_rfc.fit(x_train_os,y_train_os)
y_pred_os = os_rfc.predict(x_test)
```

```python
print(accuracy_score(y_test , y_pred_os))
print(confusion_matrix(y_test , y_pred_os))
print(classification_report(y_test , y_pred_os))
```
